```python
def detect_text(path):
    """Detects text in the file."""
    from google.cloud import vision
    import io
    client = vision.ImageAnnotatorClient()

    with io.open(path, 'rb') as image_file:
        content = image_file.read()

    image = vision.Image(content=content)

    response = client.text_detection(image=image)
    texts = response.text_annotations
    print('Texts:')

    for text in texts:
        print('\n"{}"'.format(text.description))

        vertices = (['({},{})'.format(vertex.x, vertex.y)
                    for vertex in text.bounding_poly.vertices])

        print('bounds: {}'.format(','.join(vertices)))

    if response.error.message:
        raise Exception(
            '{}\nFor more info on error messages, check: '
            'https://cloud.google.com/apis/design/errors'.format(
                response.error.message))
```

```python
def get_item_link(page, pagesize, shorder, shfiltercnt, shwinetype, shwineprice1, shwineprice2):
    # 총 페이지 갯수
    url = 'https://www.wine21.com/13_search/wine_list.html?Page={}&PageSize={}&shKeyword=&shOrder={}&shFILTERCNT={}&shWINETYPE={}&shWINEPRICE1={}&shWINEPRICE2={}#List'.format(
    page, pagesize, shorder, shfiltercnt, shwinetype, shwineprice1, shwineprice2)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    page_count = soup.select('span.current')[1].text.split('\xa0')[2]

    links = []
    # 페이지 바꿔주기
    for i in range(int(page_count)):
        url = 'https://www.wine21.com/13_search/wine_list.html?Page={}&PageSize={}&shKeyword=&shOrder={}&shFILTERCNT={}&shWINETYPE={}&shWINEPRICE1={}&shWINEPRICE2={}#List'.format(
    i, pagesize, shorder, shfiltercnt, shwinetype, shwineprice1, shwineprice2)
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        idx = soup.select('div.tit > h4 > a')
        #해당 페이지 목록 상품 링크 크롤링
        for index in idx:
            links.append("https://www.wine21.com/13_search/wine_view.html?Idx=" + index.get('idx'))
    return links
```

```python
items_df = pd.DataFrame()

for index in range(100,len(links)+100,100):
    for link in links[index-100:index]:
        data = []
        response = requests.get(link)
        dom = BeautifulSoup(response.content, 'html.parser')

        #title
        try:
            title = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.cnt > h4')[0].get_text()
        except:
            continue
        # image_link
        try:
            image_link = dom.select("#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail1 > div > img")[0].get('src')
        except:
            print('error crawling {} for {}'.format("image_link", link))
            image_link = None

        # winery
        try:
            winery = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd.winery > a')[0].text.split("\xa0")[0]
        except:
            print('error crawling {} for {}'.format("winery", link))
            winery = None

        # country, province
        try:
            country = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd.wine_area > a:nth-child(1)')[0].get_text().split(' ')[0]
        except:
            print('error crawling {} for {}'.format("country", link))
            country = None
        try:
            province = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd.wine_area > a:nth-child(2)')[0].get_text().split(' ')[0]
        except:
            print('error crawling {} for {}'.format("province", link))
            province = None

        # grape_type
        try:
            grape_type = []
            for i in range(len(dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd.variety > a'))):
                grape_type.append(dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd.variety > a')[i].get_text().strip())
        except:
            print('error crawling {} for {}'.format("grape_type", link))
            grape_type = None

        # wine type
        try:
             wine_type = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({})'.format(number('종 류')))[0].get_text().strip('\xa0')
        except:
            print('error crawling {} for {}'.format("wine type", link))
            wine_type = None

        # occasion
        try:
            occasion = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({})'.format(number('용 도')))[0].get_text().strip('\xa0')
        except:
            print('error crawling {} for {}'.format("wine type", link))
            occasion = None

        # alcohol
        try:
            alcohol = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({})'.format(number('알코올 도수')))[0].get_text()
            alcohol = mean(list(map(int, (re.findall(r'[0-9]+', alcohol)))))
        except:
            print('error crawling {} for {}'.format("alcohol", link))
            continue

        # sweetness
        try:
            sweetness = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({}) > div > img'.format(number('당 도')))[0].get('alt')[-1]
        except:
            print('error crawling {} for {}'.format("sweetness", link))
            continue

        # acidity
        try:
            acidity = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({}) > div > img'.format(number('산 도')))[0].get('alt')[-1]
        except:
            print('error crawling {} for {}'.format("acidity", link))
            acidity = None

        # body rate
        try:
            body_rate = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({}) > div > img'.format(number('바 디')))[0].get('alt')[-1]
        except:
            print('error crawling {} for {}'.format("body rate", link))
            body_rate = None

        # tannin rate
        try:
            tannin_rate = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({}) > div > img'.format(number('타 닌')))[0].get('alt')[-1]
        #frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child(22) > div > img
        except:
            print('error crawling {} for {}'.format("tannin rate", link))
            tannin_rate = None

        # food_pairing
        try:
            food_pairing = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({})'.format(number('추천 음식')))[0].get_text()
        except:
            print('error crawling {} for {}'.format("food pairing", link))
            continue

        # price
        try:
            price = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({}) > strong:nth-child(1)'.format(number('가격')))[0].get_text().split(',')[0]+'000'
        except:
            continue

        # vintage
        try:
            vintage = dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({}) > strong:nth-child(2)'.format(number('가격')))[0].get_text().split(',')[0].strip('()')
        except:
            print('error crawling {} for {}'.format("vintage", link))
            vintage = None

        # size
        try:
            size = re.findall(r'[0-9]+', dom.select('#frmMain > article > div.column_block_1.board_list.wine_view_wrap > div.column_detail2 > div.wine_info > dl > dd:nth-child({}) > strong:nth-child(2)'.format(number('가격')))[0].get_text())[-1]
        except:
            print('error crawling {} for {}'.format("size", link))
            size = None
        data = [{'title': title, 'image_link': image_link, 'winery':winery, 'country': country, 'province': province,
                 'grape_type': grape_type, 'wine_type': wine_type, 'occasion': occasion, 'alcohol': alcohol,
                 'sweetness': sweetness, 'acidity': acidity, 'body_rate': body_rate, 'tannin_rate': tannin_rate,
                 'food_pairing':food_pairing, 'price':price, 'vintage': vintage, 'size': size}]
        items_df = items_df.append(data, ignore_index=True)
```
